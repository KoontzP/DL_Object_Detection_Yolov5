```python
import os
import requests
import zipfile

# URLs for downloading the datasets
train_url = "http://images.cocodataset.org/zips/train2017.zip"
val_url = "http://images.cocodataset.org/zips/val2017.zip"
annotations_url = "http://images.cocodataset.org/annotations/annotations_trainval2017.zip"

# File names for saving the datasets
train_filename = "coco_train2017.zip"
val_filename = "coco_val2017.zip"
annotations_filename = "coco_ann2017.zip"

# Create directories to store the extracted images
train_dir = "train2017"
val_dir = "val2017"
annotations_dir = "annotations"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(annotations_dir, exist_ok=True)

# Download train dataset
print("Downloading train dataset...")
response = requests.get(train_url)
with open(train_filename, "wb") as f:
    f.write(response.content)
print("Train dataset downloaded.")

# Extract train dataset
print("Extracting train dataset...")
with zipfile.ZipFile(train_filename, 'r') as zip_ref:
    zip_ref.extractall(train_dir)
print("Train dataset extracted.")

# Download validation dataset
print("Downloading validation dataset...")
response = requests.get(val_url)
with open(val_filename, "wb") as f:
    f.write(response.content)
print("Validation dataset downloaded.")

# Extract validation dataset
print("Extracting validation dataset...")
with zipfile.ZipFile(val_filename, 'r') as zip_ref:
    zip_ref.extractall(val_dir)
print("Validation dataset extracted.")

# Download annotations
print("Downloading annotations...")
response = requests.get(annotations_url)
with open(annotations_filename, "wb") as f:
    f.write(response.content)
print("Annotations downloaded.")

# Extract annotations
print("Extracting annotations...")
with zipfile.ZipFile(annotations_filename, 'r') as zip_ref:
    zip_ref.extractall(annotations_dir)
print("Annotations extracted.")
```
