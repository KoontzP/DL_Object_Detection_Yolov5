```python
import torch # DL models
from matplotlib import pyplot as plt
import numpy as np #array transformation np-array
import cv2 #opencv, directory/file management
from IPython.display import Image, display #view images
```

```python
# model able to detect up to 81 different classes, ex. car, person, backpack etc.
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
img_path = './ford_vtti_research_04_hr_1280x720.jpg'
```

```python
display(Image(filename=img_path))
```

```python
results = model(img_path)
```

```python
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
# [xmin, ymin, xmax, ymax, confidence, class]
results.xyxy
```

```python
def play_video_with_model(video_path, model):
    capture = cv2.VideoCapture(video_path)

    while capture.isOpened():
        ret, frame = capture.read()
        
        if not ret:
            break

        results = model(frame)

        cv2.imshow('YOLO', np.squeeze(results.render()))

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    capture.release()
    cv2.destroyAllWindows()
```

```python
video_path = "./video/traffic.mp4"
play_video_with_model(video_path, model)
```
